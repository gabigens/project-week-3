```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf
```

```python
def get_brand_name ():

    url = 'https://www.lyst.com/the-lyst-index/the-lyst-index-q2-25/'
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    grid = soup.find('div', attrs = {'class':'chart-update'})
    brand_elements = grid.find_all('div', attrs = {'class':'mono-type padding-left'})

    brand_names = []
    
    for elements in brand_elements:
          clean_brand_name = elements.get_text()
          brand_names.append(clean_brand_name)

    return brand_names
```

```python
company_names = pd.read_csv('fashion_brands_parent_companies.csv')
company_names = pd.DataFrame(company_names)
company_names
```

```python
company_names_2 = company_names.drop('SourceURL', axis=1)
company_names_2
```

```python
grouped = company_names_2.groupby(['ParentCompany', 'Ticker','Brand']).count().reset_index()
grouped
```

```python
grouped = grouped.groupby(['ParentCompany', 'Ticker'])['Brand'].apply(', '.join).reset_index()
grouped
```

```python
# Extract tickers
tickers = grouped['Ticker'].unique()

# Prepare storage for data
all_data = []

# Loop through tickers and fetch data
for ticker in tickers:
    try:
        # Fetch historical data
        ticker_data = yf.Ticker(ticker).history(period='5y')
        ticker_data['Ticker'] = ticker  # Add a column with the ticker
        
        # Append the data to the list
        all_data.append(ticker_data)
        print(f"Data for {ticker} retrieved successfully.")

    except Exception as e:
        print(f"Failed to retrieve data for {ticker}: {e}")

# Concatenate all data into a single DataFrame
if all_data:
    big_data = pd.concat(all_data)
    big_data.reset_index(inplace=True)  # Reset index to have a flat structure
    print(big_data.head())
else:
    big_data = pd.DataFrame()  # Return empty DataFrame if no data
```

```python
big_data = big_data.drop(['Open', 'High', 'Low', 'Volume', 'Dividends',
       'Stock Splits', 'Adj Close'], axis=1)
big_data
```

```python
big_data.groupby(['Ticker', 'Date']).agg({'Close':['mean']})
big_data
```

```python
# Ensure the 'Date' column is in datetime format
big_data['Date'] = pd.to_datetime(big_data['Date'], utc=True)

# Extract year and month into new columns 'Year' and 'Month'
big_data['Year'] = big_data['Date'].dt.year
big_data['Month'] = big_data['Date'].dt.month
```

```python
big_data = big_data.groupby(['Ticker', 'Year', 'Month']).agg({'Close': 'mean'}).reset_index()
big_data
```
